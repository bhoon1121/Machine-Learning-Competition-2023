```python
for i in range(1,10):

    plt.figure(figsize=(12,10))
    plt.suptitle(f"Training.data #{i}")
    df=pd.read_csv(f'data/train/aligned_well_0{i}.csv')

    plt.subplot(1,4,1)
    plt.plot(df.GR,df.DEPT)
    plt.gca().invert_yaxis()
    plt.title('Gamma-ray')
    plt.xlabel('GR')
    plt.ylabel('Depth')
    plt.grid('on')

    plt.subplot(1,4,2)
    plt.plot(df.RHOB,df.DEPT)
    plt.gca().invert_yaxis()
    plt.title('Density')
    plt.xlabel('RHOB')
    plt.ylabel('Depth')
    plt.xlim([2,2.75])
    plt.grid('on')

    plt.subplot(1,4,3)
    plt.plot(df.NPHI,df.DEPT)
    plt.gca().invert_yaxis()
    plt.title('Neutron')
    plt.xlabel('NPHI')
    plt.ylabel('Depth')
    plt.grid('on')

    plt.subplot(1,4,4)
    plt.plot(df.RD,df.DEPT)
    plt.gca().invert_yaxis()
    plt.title('Resistivity')
    plt.xlabel('RD')
    plt.ylabel('Depth')
    plt.grid('on')
    plt.xlim([0,25])
    plt.tight_layout()
    
```
